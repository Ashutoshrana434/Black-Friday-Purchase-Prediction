```python
import pandas as pd

# Load the data
df = pd.read_csv('file (1).csv')

# 1. The 'Train' set (where Purchase is known)
train_full = df[df['Purchase'].notnull()]

# 2. The 'Test' set (where Purchase is unknown - what you need to predict)
test_full = df[df['Purchase'].isnull()]

print(f"Train set size: {len(train_full)}") # Should be ~550,068
print(f"Test set size: {len(test_full)}")   # Should be ~233,599
```

```python
# Save the Training set (Labeled data)
train_full.to_csv('train_data.csv', index=False)

# Save the Test set (Unlabeled data for prediction)
test_full.to_csv('test_data.csv', index=False)

print("Files saved successfully!")
```
